## Split dataset

```python
data_combine = sc.read(dataPath + 'celegans_scobj.h5ad')
```

```python
data_combine.var_names = data_combine.var.loc[:, "gene_short_name"]
```

```python
def getScObjects(scobj, item, values, n_neighbors=10, n_pcs=40):
    data_split = [scobj[scobj.obs[item] == v, :] for v in values]
    data_split = [preProcess(d, log_normalize=True) for d in data_split]
    comm_hvgs = set([g for d in data_split for g in d.var.index[d.var.highly_variable]])

    for di in range(len(data_split)):
        data_split[di].var.highly_variable = [gene in comm_hvgs for gene in data_split[di].var.index]
        data_split[di] = runScanpy(data_split[di], n_neighbors, n_pcs, regress=False)

    return (data_split)
```

```python
data_combine.obs['time.point'].unique()
```

```python
data_300, data_400, data_500 = getScObjects(data_combine, item='time.point',
                                            values=['300_minutes', '400_minutes', '500_minutes'],
                                            n_neighbors=10, n_pcs=40)
```

```python
data_300.write(dataPath + 'M300_celegans_scobj.h5ad')
data_400.write(dataPath + 'M400_celegans_scobj.h5ad')
data_500.write(dataPath + 'M500_celegans_scobj.h5ad')
```

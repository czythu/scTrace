## Split dataset

```python
dataPath = "D:/Lineage-data/Nature18-CellTagging/"
adata_orig = sc.read(dataPath+"CellTagging_adata_preprocessed_new.h5ad")
```

```python
def getScObjects(scobj, item, values, n_neighbors = 10, n_pcs = 40):
    data_split = [scobj[scobj.obs[item] == v,:] for v in values]
    data_split = [preProcess(d, log_normalize=True) for d in data_split]
    comm_hvgs = set([g for d in data_split for g in d.var.index[d.var.highly_variable]])
    
    for di in range(len(data_split)):
        data_split[di].var.highly_variable = [gene in comm_hvgs for gene in data_split[di].var.index]
        data_split[di] = runScanpy(data_split[di], n_neighbors, n_pcs, regress=False)
    
    return(data_split)
```

```python
data_1, data_2, data_3, data_4 = getScObjects(adata_orig, item = 'time_combined',
                                              values = ['2barcoding', '3barcoding-1', '3barcoding-2','3barcoding-3'],
                                              n_neighbors = 10, n_pcs = 40)
```

```python
data_1.obs["barcode_day0"][data_1.obs["barcode_day0"] == "NA"] = np.nan
data_2.obs["barcode_day0"][data_2.obs["barcode_day0"] == "NA"] = np.nan
data_3.obs["barcode_day0"][data_3.obs["barcode_day0"] == "NA"] = np.nan
data_4.obs["barcode_day0"][data_4.obs["barcode_day0"] == "NA"] = np.nan
```

```python
data_1.write(dataPath + '2barcoding_CellTagging_scobj.h5ad')
data_2.write(dataPath + '3barcoding_1_CellTagging_scobj.h5ad')
data_3.write(dataPath + '3barcoding_2_CellTagging_scobj.h5ad')
data_4.write(dataPath + '3barcoding_3_CellTagging_scobj.h5ad')
```

```python
from scTrace.pipeline import *
import gzip
import os
```

### Loop for all data

```python
dataPath_ori = 'D:/Lineage-data/NC24-ReSisTrace/'
dataSet = "NC24-ReSisTrace"
```

```python
files = sorted([os.path.join(dataPath_ori, file) for file in os.listdir(dataPath_ori)])
file_list = [files[i:i+4] for i in range(0, len(files), 4)]
label_list = ["Carboplatin_1", "Carboplatin_2", "Olaparib_1", "Olaparib_2", "NK_1", "NK_2", "Control_1", "Control_2"]
```

```python
file_list
```

```python
for files, label in zip(file_list, label_list):
    print(label)
    with gzip.open(files[1], 'rt') as f:
        cell_info = pd.read_csv(f, sep="\t", index_col=0)
    with gzip.open(files[0], 'rt') as f:
        UMI_counts = pd.read_csv(f, sep="\t", index_col=0)
    data_pre_ori = sc.AnnData(X=UMI_counts.T, obs=cell_info)
    with gzip.open(files[3], 'rt') as f:
        cell_info = pd.read_csv(f, sep="\t", index_col=0)
    with gzip.open(files[2], 'rt') as f:
        UMI_counts = pd.read_csv(f, sep="\t", index_col=0)
    data_pos_ori = sc.AnnData(X=UMI_counts.T, obs=cell_info)    
    common_genes = np.intersect1d(data_pre_ori.var.index, data_pos_ori.var.index)
    data_pre_ori = data_pre_ori[:,common_genes]
    data_pos_ori = data_pos_ori[:,common_genes]
    data_pre_ori = preProcess(data_pre_ori)
    data_pos_ori = preProcess(data_pos_ori)
    data_split = [data_pre_ori, data_pos_ori]
    comm_hvgs = set([g for d in data_split for g in d.var.index[d.var.highly_variable]])
    for di in range(len(data_split)):
        data_split[di].var.highly_variable = [gene in comm_hvgs for gene in data_split[di].var.index]
        data_split[di] = runScanpy(data_split[di], n_neighbors=10, n_pcs=40, regress=False)
    data_pre, data_pos = data_split
    
    dataSet = "NC24"
    samName = "ReSisTrace"
    run_label = dataSet + "_" + samName + "_" + label
    
    dataPath = 'D:/Lineage-data/NC24-ReSisTrace/'
    
    all_data = [data_pre, data_pos]
    all_names = ['pre', 'pos']
    
    savePath = "D:/Lineage-data/NC24-ReSisTrace/Results-reproduce/" # + run_label + "/"
    if not os.path.exists(savePath):
        os.makedirs(savePath)
    
    time = 0
    
    data_pre, data_pos = all_data[time:(time+2)]
    pre_name, pos_name = all_names[time:(time+2)]

    run_label_time = run_label + '-' + pre_name + '_' + pos_name
    print(run_label_time)
    
    cross_sim = getSimilarityMatrix(data_pre, data_pos, method = 'Pearson')
    barcodes_pre, barcodes_pos = list(data_pre.obs['prer_r_group']), list(data_pos.obs['prer_r_group'])
    cross_lin_mat, _ = getLineageMatrix(bars = barcodes_pre, bars2 = barcodes_pos)
    
    plotSimilarityCompare(cross_sim=cross_sim,
                          cross_lin_mat=cross_lin_mat,
                          title="ReSisTrace-" + label + ': ' + pre_name + ' -> ' + pos_name,
                          savePath=savePath + run_label_time + '-SimDistrComp-hist.png')

    sim_mat = getSimilarityMatrix(data_pre, data_pre, method = 'Pearson')
    within_lin_mat, _ = getLineageMatrix(bars = barcodes_pre, bars2 = barcodes_pre)
    plotSimilarityCompare(cross_sim=sim_mat,
                          cross_lin_mat=within_lin_mat,
                          title="ReSisTrace-" + label + ': ' + pre_name,
                          savePath=savePath + run_label_time + '-SimDistrComp-hist-pre.png')
    
#     scd_obj = prepareScdobj(data_pre=data_pre, data_pos=data_pos,
#                         time=time,
#                         pre_name=pre_name, pos_name=pos_name,
#                         cls_res_all=cls_res_all, clq_res_all=clq_res_all,
#                         pre_colors=pre_colors, pos_colors=pos_colors,
#                         savePath=savePath, run_label_time=run_label_time)

#     results = assignLineageInfo(scd_obj=scd_obj,
#                                 cross_lin_mat=cross_lin_mat,
#                                 savePath=savePath,
#                                 run_label_time=run_label_time)
```
